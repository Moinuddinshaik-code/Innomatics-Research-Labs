## 4. Task 2: Multi-Input Prompt System

Build a template using `topic`, `audience`, and `tone`.


```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

task_2_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

task_2_outputs = []
for case in task_2_cases:
    formatted_prompt = multi_input_template.format(**case)
    task_2_outputs.append({"inputs": case, "output": formatted_prompt})

for index, item in enumerate(task_2_outputs, start=1):
    print(f"Test Case {index}:")
    print(item["output"])
    print("-" * 80)
```

## 6. Task 4: ChatPromptTemplate System

Build a chat-based prompt system using system and user roles.


```python
role_instructions = {
    "teacher": "You are a patient teacher who explains concepts clearly for learners.",
    "interviewer": "You are an interviewer who asks thoughtful and relevant questions.",
    "motivator": "You are a motivating mentor who encourages and inspires the learner."
}

chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system", "{system_instruction}"),
    ("human", "Help me with the topic: {topic}")
])

def build_chat_prompt(role, topic):
    if role not in role_instructions:
        raise ValueError("Role must be one of: teacher, interviewer, motivator")

    messages = chat_prompt_template.format_messages(
        system_instruction=role_instructions[role],
        topic=topic
    )
    return messages

task_4_role = "teacher"
task_4_topic = "Neural Networks"
task_4_messages = build_chat_prompt(task_4_role, task_4_topic)

print("Task 4 Output:")
for message in task_4_messages:
    print(f"{message.type.upper()}: {message.content}")
```

## 7. Task 5: Input Validation Layer

Validate the `audience` and `tone` inputs before generating prompts.


```python
VALID_AUDIENCES = {"beginner", "intermediate", "expert"}
VALID_TONES = {"formal", "casual", "fun"}

def validate_inputs(audience, tone, use_defaults=False):
    audience_value = audience.strip().lower()
    tone_value = tone.strip().lower()

    if use_defaults:
        if audience_value not in VALID_AUDIENCES:
            audience_value = "beginner"
        if tone_value not in VALID_TONES:
            tone_value = "formal"
        return audience_value, tone_value

    if audience_value not in VALID_AUDIENCES:
        raise ValueError(f"Invalid audience: {audience}. Allowed values: {sorted(VALID_AUDIENCES)}")
    if tone_value not in VALID_TONES:
        raise ValueError(f"Invalid tone: {tone}. Allowed values: {sorted(VALID_TONES)}")

    return audience_value, tone_value

print("Task 5 Valid Example:")
print(validate_inputs("beginner", "fun"))

print("\nTask 5 Default Example:")
print(validate_inputs("kids", "friendly", use_defaults=True))
```

## 8. Task 6: Prompt Generator App

Create a modular function `generate_prompt(topic, audience, tone, style)` using `PromptTemplate`.


```python
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Create an {style} prompt about {topic} for {audience} using a {tone} tone"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template="Explain {topic} for {audience} in a {tone} {style} style"
    )
}

def generate_prompt(topic, audience, tone, style):
    audience_value, tone_value = validate_inputs(audience, tone, use_defaults=False)
    style_value = style.strip().lower()

    if style_value not in style_templates:
        raise ValueError("Style must be one of: teaching, interview, storytelling")

    return style_templates[style_value].format(
        topic=topic,
        audience=audience_value,
        tone=tone_value,
        style=style_value
    )

task_6_output = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print("Task 6 Output:")
print(task_6_output)
```

## 9. Task 7: Template Reusability Test

Use one template with five different inputs to prove the prompt structure is reusable.


```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

reusability_inputs = [
    {"topic": "Artificial Intelligence", "audience": "beginner", "tone": "formal"},
    {"topic": "Python Programming", "audience": "intermediate", "tone": "casual"},
    {"topic": "Data Science", "audience": "expert", "tone": "formal"},
    {"topic": "Neural Networks", "audience": "beginner", "tone": "fun"},
    {"topic": "Machine Learning", "audience": "intermediate", "tone": "casual"}
]

reusability_results = []
for item in reusability_inputs:
    prompt_text = reusable_template.format(**item)
    reusability_results.append({
        "topic": item["topic"],
        "audience": item["audience"],
        "tone": item["tone"],
        "generated_prompt": prompt_text
    })

reusability_df = pd.DataFrame(reusability_results)
reusability_df
```

## 10. Full Pipeline Flow

This final function demonstrates the expected pipeline:

`User Input -> Validation -> Prompt Template -> Dynamic Prompt Generation -> Output`


```python
def full_prompt_pipeline(topic, audience, tone, style):
    validated_audience, validated_tone = validate_inputs(audience, tone, use_defaults=False)
    final_prompt = generate_prompt(topic, validated_audience, validated_tone, style)
    return {
        "user_input": {
            "topic": topic,
            "audience": audience,
            "tone": tone,
            "style": style
        },
        "validated_input": {
            "audience": validated_audience,
            "tone": validated_tone
        },
        "generated_prompt": final_prompt
    }

pipeline_result = full_prompt_pipeline(
    topic="Large Language Models",
    audience="beginner",
    tone="casual",
    style="teaching"
)

pipeline_result
```
